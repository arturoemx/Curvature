```python
import numpy as np
import matplotlib.pyplot as plt
#enable interactive plot
%matplotlib notebook
```

```python
h=10
k=20
r=10
N=720
N2=N//2

theta=np.arange(-np.pi,np.pi,np.pi/N2)
N=len(theta)
x = h + r * np.cos(theta)#+0.05*r*np.sin(5*theta)
y = k + r * np.sin(theta)#+0.025*r*np.cos(3*theta)

xf, yf = x, y
x = x[1:N:1]
y = y[1:N:1]
```

```python
fig1, ax1 = plt.subplots(1,1)
fig1.suptitle("Circulo")
ax1.plot(x,y,'b', xf, yf, 'r')
print(x.size,xf.size)
```

```python
## Compute Curvature.

def computeCurvature(x, y, nargout):
    dxb = x[1] - x[0]
    dyb = y[1] - y[0]
    Sb = np.sqrt(dxb**2 + dyb**2)
    dxf = x[2] - x[1]
    dyf = y[2] - y[1]
    Sf = np.sqrt(dxf**2 + dyf**2)
    if Sb != 0 and Sf !=0:
        dx = 0.5 * (dxb / Sb + dxf / Sf)
        dy = 0.5 * (dyb / Sb + dyf / Sf)
        iS = 2./((Sb + Sf)*Sb*Sf)
        dx2 = iS * (Sb*dxf-Sf*dxb)
        dy2 = iS * (Sb*dyf-Sf*dyb)
    k = (dx*dy2-dy*dx2)/((dx*dx+dy*dy)**1.5)
    if nargout == 1:
        return k
    else:
        return k, Sb, dx, dy, dx2, dy2   
```

```python
N=len(x)
kc  = np.zeros(x.size)
s   = np.zeros(x.size)
dx  = np.zeros(x.size)
dy  = np.zeros(x.size)
dx2 = np.zeros(x.size)
dy2 = np.zeros(x.size)

for i in range(-1, N - 2):
    l = i + 1
    X=[x[j] for j in range(i,i+3)]
    Y=[y[j] for j in range(i,i+3)]
    kc[l], s[l], dx[l], dy[l], dx2[l], dy2[l] = computeCurvature(X, Y, 2)
l = N - 1
[kc[l], s[l], dx[l], dy[l], dx2[l], dy2[l]] = computeCurvature([x[l-1],x[l],x[0]], [y[l-1],y[l],y[0]], 2)
```

```python
fig2, ax2 = plt.subplots(1,1)
ax2.plot(x-h,y-k,'b',dx, dy,'m', dx2, dy2, 'k')

fig3, ax3 = plt.subplots(1,1)
ax3.plot(kc,  'g')
```
